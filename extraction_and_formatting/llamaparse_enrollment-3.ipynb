````python
base_prompt = """
You are a data transformation assistant specializing in financial and educational data. I will provide you with multiple pieces of extracted text from different sections of Mount Saint Mary's University’s financial report. Your task is to consolidate all this information into a single summary formatted as a JSON object that represents a table.

The table must have columns for the years: 2024, 2023, 2022, 2021, and 2020. Each row in the table should correspond to a specific metric with its value for each year.

Important instructions:
1. Units & Data Types:
   - If a value contains a "$" symbol, treat it as financial data and preserve its formatting exactly (e.g. "$25,521,495" with the dollar sign and comma separators). Financial values are typically large (often in the millions).
   - If a value includes a "%" symbol, treat it as a percentage.
   - Values without a currency or percentage symbol are assumed to be headcounts, test scores, or similar numbers. Ensure that headcount values are realistic (typically within the range 0–50,000). If you see a very large number without a "$", it is likely incorrect for headcount.
2. Data Cleaning:
   - Remove any extra notes or text inside parentheses.
   - If a metric is provided in multiple parts (for example, Graduate Enrollment broken down by college), sum the parts to yield one total value.
3. Formatting Requirements:
   - The output must be exactly valid JSON without any markdown formatting, code block markers, or extra commentary.
   - Do not include any additional explanation or text.
   - If a particular value is missing for a metric, leave it as an empty string.
4. Metrics to Include:
   Please include the following metrics in the final JSON output:
   - Undergraduate Enrollment (Headcount)
   - Graduate Enrollment (Total)
   - Total FTE Students
   - First Year Students Applications
   - First Year Students Acceptances
   - First Year Students Matriculates
   - % Enrolled (for first year students)
   - Average SAT Score (Freshman)
   - Average ACT Score (Freshman)
   - Retention Rate %
   - Graduation Rate (within 4 yrs) %
   - Graduation Rate (within 5 yrs) %
   - Tuition (for the appropriate academic year; use the starting year)
   - Room & Board (if available; include as separate values or combined as appropriate)
   - Cash Balance (as of June 30)
   - Operating Expenses (as of June 30)
   - Daily Expenses
   - Days Liquidity

The final JSON output must follow this exact structure:

{
  "Metric Name": {
    "2024": "value",
    "2023": "value",
    "2022": "value",
    "2021": "value",
    "2020": "value"
  },
  ... (other metrics) ...
}

Now, here is the extracted information from the report:
[INSERT EXTRACTED TEXT HERE]
"""
````

````python
structured_outputs = []

for i, doc in enumerate(extracted_tables):
    print(f"Processing document {i + 1}...")

    user_prompt = base_prompt + "\n\n" + doc.text[:16000]  # truncate if too large

    response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "system", "content": "You are a financial data parser and formatter."},
        {"role": "user", "content": user_prompt}
    ],
    temperature=0.2
)


    content = response.choices[0].message.content
    structured_outputs.append(content)
````

````python
import json
import re
import pandas as pd

def strip_code_block(text):
    """
    Remove markdown code block markers (e.g. ```json ... ```) from the text.
    If multiple parts exist, return the content in the first code block.
    """
    parts = text.split("```")
    if len(parts) >= 3:
        # The JSON content is usually in the second element
        return parts[1].strip()
    else:
        return text.strip()

dfs = []

for i, result in enumerate(structured_outputs):
    raw_text = result  # result is the raw GPT output (a string)

    # Strip the markdown formatting
    cleaned_text = strip_code_block(raw_text)

    try:
        data = json.loads(cleaned_text)
        df = pd.DataFrame.from_dict(data, orient='index')
        df.index.name = "Metric"
        dfs.append(df)
    except Exception as e:
        # Optionally save the raw text for further debugging
        with open(f"failed_document_{i+1}.txt", "w") as f:
            f.write(raw_text)

# Optionally, visualize the first DataFrame if any were successfully parsed
if dfs:
    display(dfs[0])
else:
    print("No documents were successfully processed.")
````
